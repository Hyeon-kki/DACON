```python
'''
< IDEA >

0. ID는 훈련에서 제외하자. (실험1)

1. Age 
    - 평균 미국 근로소득 나이대 알아봐서 나누기 (실험2)
    - 10살아래 / 60살 이상 나이대의 근로소득 파악해보기 (실험3) 

2. 

'''
```

```python
def age_category(age):
    if age<20:
        category = '0~10대'
    elif age>20 and age<30:
        category = '20대'
    elif age>30 and age<40:
        category = '30대'
    elif age>40 and age<50:
        category = '40대'
    elif age>50 and age<60:
        category = '50대'
    elif age>60 and age<70:
        category = '60대'
    else:
        category = '60대 이상'
    return category
```

```python
train_df['Age_category'] = train_df['Age'].map(age_category)
```

```python
train_df['Income'].groupby(train_df['Age_category']).mean()
```

```python
# 나이대 별 max 값
train_df['Income'].groupby(train_df['Age_category']).max()
```
